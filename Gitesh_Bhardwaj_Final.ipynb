#  Data Cleansing.
## Will drop the column which has NA more than 80 Percent.
## Will replace the value with Mean for column with less than 80 Percent.
## Drop the rows where all cell are NA.

```python
missing_values=df_loan.columns[df_loan.isnull().mean() > 0.8] # Return the values of column with more than 80 percent value NA.
df_loan.drop(missing_values, axis=1, inplace=True)
print(missing_values) # 56 Columns which has values NA more than 80 percent are dropped from the scope.
```
